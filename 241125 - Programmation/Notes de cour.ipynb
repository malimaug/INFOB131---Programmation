```python
def sum_x_first_integers(n: int):
    """spé"""

    if n == 0:
        return 0

    else:
        return n + sum_x_first_integers(n - 1)

print(sum_x_first_integers(2976))
```

```python
def fibonacci(n):
    if n <= 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(40))
```

```python
def serie(n):

    if n == 1:
        return '1'

    elif n % 2 == 0:
        return str(n/2) + ' ' + serie (n/2)

    else:
        return str(3 * n + 1) + ' ' + serie(3 * n +1)

print(serie(40))
```
